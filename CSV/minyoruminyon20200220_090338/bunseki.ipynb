```python
%matplotlib inline
def showgraph(name, title, xlabel, ylim):
    import numpy as np
    import matplotlib.pyplot as plt

    f=sorted([i.split() for i in open(name).readlines()], key=lambda x: x[1])
    x,y=[],[]
    for i in f:
        x.append(i[1])
        y.append(int(i[0]))

    plt.figure(figsize=(10, 10), dpi=100)
    plt.xlim(0,len(x)-1)
    plt.ylim(0,ylim)
    plt.grid(color='gray')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('tweet(s)')
    plt.plot(x, y)
    plt.xticks(rotation=90)
showgraph(
    "zikanbetsu",
    """shellgeibot tweet by hours
    877833072426663936(20170622) - 1200423686248923137(20191129)""",
    "HH",
    4000
)
showgraph(
    "tsukibetsu",
    """shellgeibot tweet by months
    877833072426663936(20170622) - 1200423686248923137(20191129)""",
    "HH",
    8000
)
```
